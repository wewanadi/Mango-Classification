```python
import csv

with open('./Data/Train2/train.csv','r') as csvfile:
    rows = csv.reader(csvfile)
    train_info = list(rows)
with open('./Data/Train2/dev.csv','r') as csvfile:
    rows = csv.reader(csvfile)
    dev_info = list(rows)  
    
# print (train_info)
print (dev_info)
```

```python
import cv2

image_path = r'./Data/Train2'

X_train, X_dev, y_train, y_dev = list(), list(), list(), list()

for info in dev_info[1:]:
    img = cv2.imread(image_path + '/Dev/' + info[0])
    img = img[:, :, [2,1,0]]
    X_dev.append(img)
    y_dev.append([ord(info[1])-ord('A'),dev_info[2],dev_info[3],dev_info[4],dev_info[5]])
    
for info in train_info[1:]:
    img = cv2.imread(image_path + '/Train/' + info[0])
    img = img[:, :, [2,1,0]]
    X_train.append(img)
    y_train.append([ord(info[1])-ord('A'),train_info[2],train_info[3],train_info[4],train_info[5]])
```

```python
import numpy as np

np.save('./X_train2',np.array(X_train))
np.save('./X_dev2',np.array(X_dev))
np.save('./y_train2',np.array(y_train))
np.save('./y_dev2',np.array(y_dev))
```

```python
import csv

with open('./Data/C1-P1_Train Dev_fixed/train.csv','r') as csvfile:
    rows = csv.DictReader(csvfile)
    rows = list(rows)

# Get label
mango_label = list()
for row in rows:
    if row['label'] == 'A':
        mango_label.append(1)
    
    if row['label'] == 'B':
        mango_label.append(2)
        
    if row['label'] == 'C':
        mango_label.append(3)
```

```python
import os
import cv2

# Get image name
img_names = os.listdir('./Data/C1-P1_Test')
img_names.sort()

# Get image
mango = list()
print (img_names)
for name in img_names:
    img = cv2.imread('./Data/C1-P1_Test/{}'.format(name))
    img = img[:, :, [2,1,0]]
    mango.append(img)
```

```python
import numpy as np

img_names = np.asarray(img_names)
np.save('./mango_test_name.npy',img_names)
print (img_names)
```

```python
import numpy as np

# store image to npy
mango_img = np.asarray(mango)
np.save('./mango_train_data.npy',mango_img)
mango_label = np.asarray(mango_label)
np.save('./mango_train_label.npy',mango_label)
```

```python
import numpy as np
# load data
mango_data = np.load('./mango_train_data.npy',allow_pickle=True)
mango_label = np.load('./mango_train_label.npy',allow_pickle=True)
```

```python
mango_cut = list()
for mango in mango_data:
    center = [int(np.shape(mango)[0]/2),int(np.shape(mango)[1]/2)]
    mango_cut.append(mango[center[0]-400:center[0]+400,center[1]-400:center[1]+400])

np.save('./mango_data_cut.npy',mango_cut)
```

```python
res_mango = list()
for mango in mango_data:
    large_size = max(int(np.shape(mango)[0]),int(np.shape(mango)[1]))
    mango = cv2.copyMakeBorder(mango,int((large_size-np.shape(mango)[0])/2),int((large_size-np.shape(mango)[0])/2)
                           ,int((large_size-np.shape(mango)[1])/2),int((large_size-np.shape(mango)[1])/2)
                           ,cv2.BORDER_REPLICATE)
    res_mango.append(cv2.resize(mango,(456,456),interpolation = cv2.INTER_CUBIC))    
np.save('./res_train_mango(456).npy',res_mango)
```
